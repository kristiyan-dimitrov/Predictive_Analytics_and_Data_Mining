## Preparing Data for Clustering

```python
clusterData = pd.DataFrame()
clusterData.head()
```

```python
clusterData['phone'] = np.log(cable.phone+1)
clusterData['internet'] = cable.internet
clusterData['video'] = np.log(cable.video+1)
clusterData.head()
```

```python
clusterData.describe()
```

## Trying some clustering

```python
# Initialize KMeans
kmeans_3 = KMeans(n_clusters=3, random_state=42, n_init = 100)
kmeans_3.fit(clusterData)
```

```python
kmeans_3.labels_
```

```python
kmeans_3.cluster_centers_
```

```python
kmeans_3.inertia_
```

```python
clusterData.head()
```

```python
inertia = []

for k in range(2,10):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init = 100)
    kmeans.fit(clusterData)
    inertia.append(kmeans.inertia_)
inertia
```

```python
type(inertia)
```

```python
inertiaSeries = pd.Series(inertia)
inertiaSeries.index = range(2,10)
```

```python
ax = inertiaSeries.plot()
ax.set_title('SSE as a function of k = Number of Clusters')
ax.set_ylabel('SSE')
ax.set_xlabel('Number of Clusters')
ax.figure.savefig('SSE ~ Number of clusters.png')
```

It seems k = 3 is the best solution!
