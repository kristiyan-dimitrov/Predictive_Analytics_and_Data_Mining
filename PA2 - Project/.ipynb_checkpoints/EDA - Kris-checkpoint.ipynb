```python
LatAndLongdf = df[['LATITUDE','LONGITUDE']]
```

This creates a GeoDataFrame based on X & Y! Very useful!

```python
gdf = gpd.GeoDataFrame(
    LatAndLongdf, geometry=gpd.points_from_xy(LatAndLongdf.LONGITUDE, LatAndLongdf.LATITUDE))
```
